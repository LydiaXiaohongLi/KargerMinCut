```python
import numpy as np
import math
import random
import copy
graph_al = []
with open('Data\kargerMinCut.txt') as file:
    for line in file:
        graph_al.append(line.split())
```

```python
class Karger(object):

    def __init__(self, graph_al):
        self.graph_al_bk=copy.deepcopy(graph_al)
        self.graph_al = copy.deepcopy(graph_al)
        self.node_list=set([y for x in graph_al for y in x])
        self.mincut=0
    
    def get_Karger_mincut(self):
        
        for k in range(len(self.node_list) * len(self.node_list)):
            self.graph_al=copy.deepcopy(self.graph_al_bk)
            self.node_list=set([y for x in self.graph_al for y in x])
            random.seed(k)           
            self.node_fuse()
        return
    
    def node_fuse(self):
        #basecase, if only two distinct nodes left, count the total number of edges left
        if len(self.node_list)==2:
            edge_count=sum([len(x)-1 for x in self.graph_al])/2
            if (self.mincut>edge_count) | (self.mincut ==0):
                self.mincut=edge_count 
        else:
            #Choose any two nodes from distinct nodes
            node1, node2=random.sample(self.node_list,2)  
            self.node_list.remove(node2)
            #update the original graph_al, for all nodes:
            for i in range(len(self.graph_al)):
                #if node is node1, or node2, remove the adjacency node in the list if it is node1 or node2
                if self.graph_al[i][0]==node1:
                    self.graph_al[i] = [self.graph_al[i][j] for j in range(len(self.graph_al[i])) if self.graph_al[i][j] != node2]
                elif self.graph_al[i][0]==node2:
                    self.graph_al[i] = [self.graph_al[i][j] for j in range(len(self.graph_al[i])) if self.graph_al[i][j] != node1]
                    self.graph_al[i][0]=node1
                #for all other nodes, any occurances of node 2, replace it with node 1
                else:
                    for j in range(len(self.graph_al[i])-1):
                        if self.graph_al[i][j+1]==node2:
                            self.graph_al[i][j+1]=node1
            self.node_fuse()
        return
```
